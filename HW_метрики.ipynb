# **Метрики**

```python
import numpy as np
```

```python
state = np.random.RandomState(seed=42)
homework_data = state.gamma(shape=10, scale=1, size=10000)

print(f'MSE {homework_data.mean():.8f}')
print(f'MAE {np.median(homework_data):.8f}')

homework_data.sort()
weights = 1 / homework_data
cumsum = weights.cumsum()
cutoff = weights.sum() / 2.0
print(f'MAPE {homework_data[cumsum >= cutoff][0]:.8f}')
```

```python
state = np.random.RandomState(seed=42)
homework_data = state.gamma(shape=10, scale=1, size=500000000)

print(f'MSE {homework_data.mean():.8f}')
print(f'MAE {np.median(homework_data):.8f}')

homework_data.sort()
weights = 1 / homework_data
cumsum = weights.cumsum()
cutoff = weights.sum() / 2.0
print(f'MAPE {homework_data[cumsum >= cutoff][0]:.8f}')
```

```python
state = np.random.RandomState(seed=42)
pred_values = state.gamma(shape=10, scale=1, size=10000)
target_values = pred_values + state.normal(loc=0, scale=1, size=10000)

n = pred_values.shape[0]
roc_auc = 0
for i in range(0, n):
    for j in range(i + 1, n):
        roc_auc += int((pred_values[i] < pred_values[j]) == (target_values[i] < target_values[j]))

roc_auc *= 2 / ((n * (n - 1)))
print(f"ROC_AUC {roc_auc:.8f}")
```
